## Constructing Classes 

```python
# Defining a class 
class Point():
  pass
```

Now we've created a class called `Point`. Above is the convention that how we define a class. 

The next step we'll create instances from our class. 

```python
# Creating instances of a class 
point1 = Point()
point2 = Point()
```

Think about class as a factory, and the things produces by the factory (class) are called **instances**. We've created two instances namely `point1` and `point2`. 

Eventhough it looks similar like calling a function, but it does more than that. It's creating a new instance of the `Point` class.

Screenshot 2021-04-12 at 11.22.31 PM.png

So everytime we create an instance of our class `Point()` it means it creates each individual dots on the above plot. 

```python
# Let's print our point1
print(point1) , print(point2)
```

```python
# Are they equal? 
point1 == point2
```

Seems `point1` and `point2` are different instances which has nothing to do with eachother, except the fact they're from the class `Point`.

There's something called **instance variable**, let's look into them! 

```python
# Creating an instance variable 
point1.x = 10 
point2.x = 5
```

```python
# Printing out our instance variable 
print(f'Instance variable of point1: {point1.x}')
print(f'Instance variable of point2: {point2.x}')
```

Screenshot 2021-04-12 at 11.42.34 PM.png

Instance variable are something that lives inside a instance.

So far we've seen how to define a class, what is instance and instance variables. Now it's time for methods. 

**Methods** are like a function which belonsg to a class.

```python
# Defining a method inside our class 
class Point():
  # Creating a method which gets the value of x
  def get_x(self):
    return self.x

# Creating instance of a class
point3 = Point()
point4 = Point()

# Creating instance variable 
point3.x = 200 
point4.x = 333 

# Calling our method 
print(f'Value of the method "get_x": {point3.get_x()}')
```

Alrighty!! What really happened up there? 
- First we created a class called `Point`. 
- Then we created a method called `get_x` inside our class. 
- We created a instance of our class namely `point3` and `point4`
- Created instance variables. 
- Finally called our method. Which returns 200, but how? 

Whenever we're calling a method, we call a particular instance variable of that method. 

Our method job is to grab and return `x` for us. And `x` is our instance variable which we created above and we create a method to access it. 

**What's with the `self`?**

The keyword `self` takes our instances `point1` and `point2` automatically, we don't wanna pass in the instances in our methods. 

`self.x` --> `point1.x` which is 200

But we should not set or intialize our instance variables  like `point1.x = 5` and so on. There is better way of doing this, which is by using a **constructor**.

**Constructor** (or) an *intializer method* is a special method which is used to initialize instances and including instance variables.

We can define by using a dunderscor method `__init__`. 

```python
# Constructing our class using a Constructor 

class Point1():
  # Using our dunderscore method (or) constructor to initialize our instance variables
  def __init__(self , x , y):
    self.x = x 
    self.y = y 

  # Methood for grabbing x
  def getX(self):
    return self.x

  # Method for grabbing y
  def getY(self):
    return self.y

# Creating instance our class + passing in the values of our instance variables right inside! 
point1 = Point1(100 , 200)
point2 = Point1(0.5 , 0.25)

# Now accessing our x and y (instance variables)
print(f'Instance variable x of point1 instance: {point1.x}')
print(f'\nInstance variable y of point1 instance: {point1.y}')

# Using our method to grab x and y 
print(f'\nOur getX function grabbed: {point1.getX()} of point1 instance')
```

Wow! wasn't that cool? Rather than setting instance variable by `point1.x = 5` we can now initialize our instance variables right inside the **constructor**. 

> Later on, you will see classes where the __init__ method does more than just save parameters as instance variables. For example, it might parse the contents of those variables and do some computation on them, storing the results in instance variables. It might even make an Internet connection, download some content, and store that in instance variables.
